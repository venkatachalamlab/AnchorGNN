```python

from scipy.ndimage import gaussian_filter1d
from scipy.interpolate import interp1d
plt.style.use(['science','no-latex'])
plt.figure(figsize=(4, 3),dpi = 300)
for i in range(len(train_history_path_list[0:])):

    train_history_path = train_history_path_list[i]
    df = pd.read_csv(train_history_path + '/model/train_history.txt', sep=" ", header=None)
    df.columns = ["epoch","loss","acc","precision","recall","FPR","f_score","TPR"]

    x = df['epoch'].values
    y = df['recall'].values

    
    # sort just in case
    sorted_idx = np.argsort(x)
    x = x[sorted_idx]
    y = y[sorted_idx]
    
    # define safe interpolation range
    x_new = np.linspace(x.min(), 250 - 1e-6, 500)  # stay within bounds

    f_interp = interp1d(x, y, kind='nearest', fill_value="extrapolate")
    y_interp = f_interp(x_new)
    if i==0:
        y_interp += np.random.normal(scale=1e-3, size=y_interp.shape)
    
    y_smooth = gaussian_filter1d(y_interp, sigma=2)


    if i==2:
        plt.plot(x_new, y_smooth, label=legend[i],color='red')
    else:
    
    
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.xlim(0,250)  
    plt.grid(True)

    legendax=plt.legend(loc="upper right")
    frame = legendax.get_frame()
    frame.set_facecolor('white')
    frame.set_edgecolor('black')
    
plt.tight_layout()
plt.savefig("./paper_plots/F1_score_epoch.png",dpi=300, transparent=True)
plt.show()
```

### Locating the best linking Probabality score

Run the MatchPartial.model_evaluation.py file to obtain the following 'match_acc_list.h5'

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import itertools
import seaborn as sns
import scienceplots

def plot_threshold_vs_accuracy_and_matches(dataset_path):
    # Input setup
    dataset = Path(dataset_path)
    h5_files = [
        dataset / 'match_acc_list1.h5',
        dataset / 'match_acc_list2.h5',
        dataset / 'match_acc_list3.h5'
    ]
    method_labels = {
        'match_acc_list1': 'Match Method 1',
        'match_acc_list2': 'Match Method 2',
        'match_acc_list3': 'Match Method 3'
    }

    # Fixed styles
    colors = itertools.cycle(['green', 'blue', 'orange'])
    markers = itertools.cycle(['o', 's', '^'])
    linestyles = itertools.cycle(['--', '-.', ':'])

    color_map, marker_map, line_map = {}, {}, {}
    for file in h5_files:
        stem = file.stem
        color_map[stem] = next(colors)
        marker_map[stem] = next(markers)
        line_map[stem] = next(linestyles)

    # Plot style
    plt.style.use(['science', 'no-latex'])
    fig, ax1 = plt.subplots(figsize=(7, 5),
                            dpi = (600)
                           )

    # Plot Accuracy on ax1
    for file in h5_files:
        stem = file.stem
        df = pd.read_hdf(file, key='data')
        # print(df)
        ax1.plot(df['threshold'],  df['num_matched'] / 1057,
                 marker=marker_map[stem], linestyle='-', linewidth=0.8,
                 color=color_map[stem], label=f'{method_labels[stem]} (Accuracy)', markersize=6)

    ax1.set_ylabel('Number of Matched ($n$)', color='black', fontsize=14)
    ax1.set_xlabel('GNN Match Probability Threshold ($p$)', fontsize=14)
    ax1.tick_params(axis='y', labelcolor='black')
    ax1.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
 
    
    # Plot Matched count on ax2
    ax2 = ax1.twinx()
    for file in h5_files:
        stem = file.stem
        df = pd.read_hdf(file, key='data')
        ax2.plot(df['threshold'], df['acc'],
                 marker=marker_map[stem], linestyle=line_map[stem], linewidth=1,
                 color=color_map[stem], label=f'{method_labels[stem]} (Matched)', markersize=6)
    
    ax2.set_ylabel('Accuracy ($A$)', color='red', fontsize=14)
    ax2.tick_params(axis='y', labelcolor='red')
    ax2.set_ylim(0.92, 1)
    
    # Color-coded twin legends

    leg1 = ax1.legend(loc='upper left', title='Matched Count', fontsize=9)
    leg2 = ax2.legend(loc='upper right', title='Accuracy', fontsize=9)
    leg1.get_title().set_color('black')
    leg2.get_title().set_color('red')
    ax1.add_artist(leg1)

    plt.title('Threshold vs Accuracy and Matched Count', fontsize=18)
    plt.tight_layout()
    plt.show()
    plt.savefig('./paper_plots/threshold_vs_accuracy_matched_rgb.png', transparent=True)


# Call it
plot_threshold_vs_accuracy_and_matches('dataset/ZM9624_masked_anchor')
```
